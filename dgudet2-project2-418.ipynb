```python
import pandas as pd 
df = pd.read_csv('./Traffic_Crashes_Crashes.csv')
df = df.drop(columns=['CRASH_RECORD_ID'])                                    #Drop the column 'CRASH_RECORD_ID'
#df['INJURIES_NO_INDICATION'] = pd.to_numeric(df['INJURIES_NO_INDICATION'])
#df   #original data     

 
```

```python
#Lets separate CRASH_DATE in to DATE and TIME 
df[['DATE','TIME']] = df.CRASH_DATE.str.split(expand=True)         #split the DATE_TIME into DATE and TIME 
df
```

```python
df = df.drop(columns=['CRASH_DATE'])                #move the date time year to the front
df
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df['YEAR'] = df['DATE'].dt.year                               #find the year
df
```

```python
df[['TRAFFIC_CONTROL_DEVICE','DEVICE_CONDITION','WEATHER_CONDITION', 'LIGHTING_CONDITION',
     'FIRST_CRASH_TYPE', 'TRAFFICWAY_TYPE', 'ROADWAY_SURFACE_COND',
     'ROAD_DEFECT', 'CRASH_TYPE', 'INTERSECTION_RELATED_I', 'HIT_AND_RUN_I',
     'DAMAGE', 'DATE_POLICE_NOTIFIED', 'PRIM_CONTRIBUTORY_CAUSE']] = df[['TRAFFIC_CONTROL_DEVICE','DEVICE_CONDITION','WEATHER_CONDITION', 'LIGHTING_CONDITION',
     'FIRST_CRASH_TYPE', 'TRAFFICWAY_TYPE', 'ROADWAY_SURFACE_COND',
     'ROAD_DEFECT', 'CRASH_TYPE', 'INTERSECTION_RELATED_I', 'HIT_AND_RUN_I',
     'DAMAGE', 'DATE_POLICE_NOTIFIED', 'PRIM_CONTRIBUTORY_CAUSE']].fillna('MISSING') # Specific columns 

df["HIT_AND_RUN_I"].head(15)          
```
